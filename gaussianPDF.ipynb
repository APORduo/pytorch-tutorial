```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import scipy.stats as st
import seaborn
```

```python
def plot_gaussian_distribution(means, stds, n_basesample = 2000,alpha_line = 0.5,alpha_fill = 0.6):
  title_fontsize = 20
  x_label_fontsize = y_label_fontsize = 15
  plt.figure(figsize=(8,6))

  colors = ['r','g','purple','olive','gray','gold','teal','b','crimson','orange','deepskyblue','orchid']

  plt.style.use('seaborn-v0_8') # 紫色背景

  for i in range(len(means)):
    s = np.random.normal(means[i],stds[i],n_basesample)
    s_fit = np.linspace(s.min(),s.max())

    plt.plot(s_fit,st.norm(means[i],stds[i]).pdf(s_fit),lw=1,c= colors[i],alpha = alpha_line)
    plt.fill_between(x=s_fit,y1=0,y2 = st.norm(means[i],stds[i]).pdf(s_fit),facecolor = colors[i],alpha = alpha_fill)


  plt.title("PDF",fontsize = title_fontsize)
  plt.xlabel("xlabel",fontsize = x_label_fontsize)
  plt.ylabel("ylabel",fontsize = y_label_fontsize)




```

```python
means = [2,4,5]
stds = [0.8,1,0.5]
plot_gaussian_distribution(means,stds)
```
